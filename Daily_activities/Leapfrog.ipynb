```python
%matplotlib inline
from math import sin, pi,cos
from numpy import linspace, array
from pylab import plot, show, axes
```

```python
g = 9.81 # m/s
l = 1
# define your function for your ODEs here
# Reference pp 350 in your book if you need
def f(r):
    theta = r[0]
    omega = r[1]
    dtheta = omega
    domega = -g/l * sin(theta)
    return array(domega)
```

```python
# define your fourth-order Runge-Kutta here
# reference pp 345 in your book for this
T = linspace(0,100,1000)
theta0 = pi/4
omega0 = 0.0
xs = []
ys = []
r = array([theta0,omega0],float)
print(r)
xs.append(l*sin(r[0]))
ys.append(-l*cos(r[0]))
print(xs, ys)
h = .1
for t in T:
    k1 = h*f(r)
    #print(k1)
    k2 = h*f(r+0.5*k1)
    k3 = h*f(r+0.5*k2)
    k4 = h*f(r+k3)
    r+= (k1+k2+2*k3+k4)/6
    xs.append(l*sin(r[0]))
    ys.append(-l*cos(r[0]))
```

```python
plot(xs,ys,'o')
plot(0,0,'o')
axes().set_aspect("equal")
show()
```

```python
print(len(T), len(xs))
plot(T,xs[:-1])
show()
plot(T,ys[:-1])
```
